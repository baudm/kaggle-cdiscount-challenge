```python
import numpy as np
from skimage.data import imread
import cv2
import matplotlib.pyplot as plt
import bson
import  time
import io
%matplotlib inline
```

```python
f=open('input/train.bson', 'rb')
data = bson.decode_file_iter(f)
for i in range(200):
    d = next(data)
f.close()
```

```python
p = d['imgs'][0]['picture']
```

```python
s = time.time()

b=imread(io.BytesIO(p))
print((time.time() - s)*1000,'ms')
s = time.time()
a=cv2.imdecode(np.fromstring(p, 'uint8'), cv2.IMREAD_COLOR)
a=cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
print((time.time() - s)*1000,'ms')
```

```python
a=cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
z=plt.imshow(a)
```
